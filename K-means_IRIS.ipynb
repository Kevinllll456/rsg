# K - means
* Calcula a distância (ex. euclidiana) entre objetos
* Divide os objetos em aglomerados (é definido pelo usuário)
* Cada aglomerado é definido por um centroide (centro geométrico)
* Cada objeto é interativamente associado ao centroide mais próximo

```python
from sklearn.cluster import KMeans
```

## Com 2 aglomerados

```python
#Com 2 aglomerados
Kcluster2 = KMeans(n_clusters = 2, random_state=5) #inicializa o Kmeans com 2 aglomerados

Kcluster2.fit(iris.data)

Klabels2 = Kcluster2.labels_               #rótulo dos aglomerados determinados
centroids2 = Kcluster2.cluster_centers_        #coordenadas do centroide dos aglomerados determinados
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

fig.suptitle('Características da sépala')

scattering = axes[0].scatter(iris.data[:, 0], iris.data[:, 1], c=Klabels2, s=35, edgecolor="None", cmap="plasma")
axes[0].scatter(centroids2[:, 0], centroids2[:, 1], c=[0, 1], marker='^', s=80, cmap="Dark2")
axes[0].legend(scattering.legend_elements()[0], ["Grupo 1", "Grupo 2"], loc="lower right", title="Clusters")
axes[0].set_xlabel('Comprimento da sépala')
axes[0].set_ylabel('Largura da sépala');
axes[0].set_title('Grupos do modelo')

scattering1 = axes[1].scatter(iris.data[:,0], iris.data[:,1], c = iris.target, s = 35, edgecolor = "None", cmap = "plasma")
axes[1].set_xlabel('Comprimento da sépala')
axes[1].set_ylabel('Largura da sépala');
axes[1].legend(scattering1.legend_elements()[0], iris.target_names, loc="lower right", title="Classes")
axes[1].set_title('Grupos reais')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

fig.suptitle('Características da pétala')

scattering = axes[0].scatter(iris.data[:, 2], iris.data[:, 3], c=Klabels2, s=35, edgecolor="None", cmap="plasma")
axes[0].scatter(centroids2[:, 2], centroids2[:, 3], c=[0, 1], marker='^', s=80, cmap="Dark2")
axes[0].legend(scattering.legend_elements()[0], ["Grupo 1", "Grupo 2"], loc="lower right", title="Clusters")
axes[0].set_xlabel('Comprimento da pétala')
axes[0].set_ylabel('Largura da pétala');
axes[0].set_title('Grupos do modelo')

scattering1 = axes[1].scatter(iris.data[:,0], iris.data[:,1], c = iris.target, s = 35, edgecolor = "None", cmap = "plasma")
axes[1].set_xlabel('Comprimento da pétala')
axes[1].set_ylabel('Largura da pétala');
axes[1].legend(scattering1.legend_elements()[0], iris.target_names, loc="lower right", title="Classes")
axes[1].set_title('Grupos reais')
plt.tight_layout()
plt.show()
```

O chute para 3 aglomerados parece separar bem as classes. Porém, sem conhecimento prévio, outros valores também poderiam parecer razoáveis.

A função de custo mínimo (\<nomeVariavel\>.inertia_) calcula a soma das distâncias ao quadrado das amostras para os centroides. Ou seja, quanto menor a inércia, mais coesos são os grupos.

A tendência é que quanto mais clusters houverem, menor é a distância entre seus elementos e os centroides. Então para avaliar o melhor número de clusters é preciso usar outros métodos mais elaborados que não serão vistos aqui no curso. Caso queira investigar, procure por método do cotovelo e coeficiente de silhueta.

```python
cost2 = Kcluster2.inertia_    #calcula o custo mínimo dos números de aglomerados
cost3 = Kcluster3.inertia_
print(cost2, cost3)
```

```python
# Com n aglomerados:
for i in range(1,20):
    Kcluster = KMeans(n_clusters = i, random_state=5)
    Kcluster.fit(iris.data)
    
    Klabels = Kcluster.labels_               #rótulo dos aglomerados determinados
    centroids = Kcluster.cluster_centers_        #coordenadas do centroide dos aglomerados determinados
    print(i, Kcluster.inertia_)
```

Como o K-means usa a distância euclidiana para aglomerar os pontos, a magnitude de cada característica desempenha um forte efeito no resultado. É útil "pesar" as características antes de fazer os aglomerados.

```python
print("feature\t\t\tmean\tstd\tmin\tmax")
for featnum, feat in enumerate(iris.feature_names):
    print("{:s}\t{:.2f}\t{:.2f}\t{:.2f}\t{:.2f}".format(feat, np.mean(iris.data[:,featnum]),
                                                        np.std(iris.data[:,featnum]), np.min(iris.data[:,featnum]),
                                                        np.max(iris.data[:,featnum])))
    #imprime a média, o desvio padrão, o valor mínimo e o valor máximo das características, usando o numpy

print("\n O comprimento da pétala (petal length) apresenta a maior variação, o que terá um grande efeito na aglomeração.")
```

## Normalização dos dados (préprocessamento)

```python
from sklearn.preprocessing import StandardScaler
#Padroniza as caracterīsticas removendo a média e "escalona" para uma variação unitária
scaler = StandardScaler().fit(iris.data)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

fig.suptitle('Diferença na distribuição dos dados')

bp = axes[0].boxplot(iris.data)
axes[0].set_ylabel('Valor')
axes[0].set_xlabel('Caracteristicas')
axes[0].set_xticklabels(iris.feature_names)
axes[0].set_title('Sem normalização')


bp = axes[1].boxplot(scaler.transform(iris.data))
axes[1].set_ylabel('Valor')
axes[1].set_xlabel('Caracteristicas')
axes[1].set_xticklabels(iris.feature_names)
axes[1].set_title('Com normalização')

fig.show()
```

```python
Kcluster_new = KMeans(n_clusters = 3, random_state=5)

Kcluster_new.fit(scaler.transform(iris.data))   #faz os aglomerados
centroids_new = Kcluster_new.cluster_centers_ # está no espaço dos dados transformados
centroids_original = scaler.inverse_transform(centroids_new) # transforma de volta ao espaço original
print(centroids_new)
print(centroids_original)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

scattering = axes[0].scatter(iris.data[:,0], iris.data[:,1], c = Kcluster_new.labels_, s = 35, edgecolor = "None", cmap = "plasma")
axes[0].scatter(centroids_original[:, 0], centroids_original[:, 1], c=[0, 1, 2], marker='^', s=80, cmap="Dark2")
axes[0].legend(scattering.legend_elements()[0], ["Grupo 1", "Grupo 2", "Grupo 3"], loc="lower right", title="Clusters")
axes[0].set_xlabel('Comprimento da sépala')
axes[0].set_ylabel('Largura da sépala');
axes[0].set_title('Grupos do modelo com normalização')

scattering1 = axes[1].scatter(iris.data[:,0], iris.data[:,1], c = Kcluster3.labels_, s = 35, edgecolor = "None", cmap = "plasma")
axes[1].scatter(centroids3[:, 0], centroids3[:, 1], c=[0, 1, 2], marker='^', s=80, cmap="Dark2")
axes[1].set_xlabel('Comprimento da sépala')
axes[1].set_ylabel('Largura da sépala');
axes[1].legend(scattering1.legend_elements()[0], ["Grupo 1", "Grupo 2", "Grupo 3"], loc="lower right", title="Clusters")
axes[1].set_title('Grupos do modelo sem normalização')


scattering2 = axes[2].scatter(iris.data[:,0], iris.data[:,1], c = iris.target, s = 35, edgecolor = "None", cmap = "plasma")
axes[2].set_xlabel('Comprimento da pétala')
axes[2].set_ylabel('Largura da pétala');
axes[2].legend(scattering2.legend_elements()[0], iris.target_names, loc="lower right", title="Classes")
axes[2].set_title('Grupos reais')
plt.tight_layout()
plt.show()
```
